## Move data from json files to Dataframe

```python
season_dfs = []
```

```python
for i in range(1990,2023):
    season_dfs.append(pd.DataFrame(json.load(open('data/' + str(i) + '_season.json'))['standings']))
    
```

```python
season_dfs[0]
```

## Data Processing/Cleaning
Removing goals and behinds columns

```python
for i in range(0,len(season_dfs)):
    season_dfs[i].drop(columns=['goals_for','goals_against','behinds_for','behinds_against'], inplace=True)
    
```

## Load data to Postgres

```python
engine = create_engine('postgresql://postgres:' + password + '@localhost:5432/mydataproj')
```

```python
for j in range(0,33):
    season_dfs[j].to_sql("afl_test_table_" + str(j+1990), engine)
```
